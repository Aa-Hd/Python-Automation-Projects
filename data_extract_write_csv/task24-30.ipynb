```python
import xlwings as xw
import pandas as pd
```

```python
app=xw.App(visible=True)        #opens new workbook   #visible means changes are seen

wb1=app.books('Book1')
```

```python
df=pd.read_csv(r"D:\tasks\matches_csv.csv")    #load data 
```

```python
for i in range(6):          #adding 6 sheets
    wb1.sheets.add()
```

24.	Fetch Matches for 'Mumbai Indians': How would you fetch the matches where 'Mumbai Indians' played as 'team1'?

```python
Mumbai_indians=df[df['team1']=='Mumbai Indians']
sheet1=wb1.sheets['Sheet1']
#sheet1.range('A1').value='Mumbai_indians'
sheet1.range('A1').expand().value=Mumbai_indians
sheet1.range('A:B').delete()
sheet1.autofit()
```

25.	Minimum Win by Runs: What is the smallest margin of victory (in runs) for a match in the dataset?

```python
min_win_runs=min(df['win_by_runs'])
sheet2=wb1.sheets['Sheet2']
sheet2.range('A1').value='min_win_runs'
sheet2.range('B1').expand().value=min_win_runs
#sheet2.range('A:B').delete()
#sheet2.autofit()
```

26.	Maximum Win by Wickets: What is the highest number of wickets by which a team won a match?

```python
max_win_wick=max(df['win_by_wickets'])
sheet3=wb1.sheets['Sheet3']
sheet3.range('A1').value='max_win_wick'
sheet3.range('B1').expand().value=max_win_wick
#sheet2.range('A:B').delete()
#sheet2.autofit()
```

27.	Count of Non-Null Values in 'City': How many matches have a non-null 'city' value?

```python
city_notnull=df['city'].count()
sheet4=wb1.sheets['Sheet4']
sheet4.range('A1').value='city_notnull'
sheet4.range('B1').expand().value=city_notnull
#sheet2.range('A:B').delete()
#sheet2.autofit()
```

28.	Count Unique Teams in 'team1': How many different teams are there in the 'team1' column?

```python
unique_team1=df['team1'].nunique()
sheet5=wb1.sheets['Sheet5']
sheet5.range('A1').value='unique_team1'
sheet5.range('B1').expand().value=unique_team1
#sheet2.range('A:B').delete()
#sheet2.autofit()
```

29.	Total Runs Scored: What is the total number of runs scored in all matches combined?

```python
total_runs=df['win_by_runs'].sum()
sheet6=wb1.sheets['Sheet6']
sheet6.range('A1').value='total_runs'
sheet6.range('B1').expand().value=total_runs
#sheet2.range('A:B').delete()
#sheet2.autofit()
```

30.	Mean Win by Runs: What is the average margin of victory (in runs) for all matches?

```python
mean_win_runs=df['win_by_runs'].mean().__round__(1)
sheet7=wb1.sheets['Sheet7']
sheet7.range('A1').value='mean_win_runs'
sheet7.range('B1').expand().value=mean_win_runs
```
