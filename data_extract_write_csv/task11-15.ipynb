```python
import xlwings as xw
import pandas as pd
```

```python
app=xw.App(visible=True)        #opens new workbook   #visible means changes are seen

wb1=app.books('Book1')
```

```python
df=pd.read_csv(r"D:\tasks\matches_csv.csv")    #load data 
```

```python
for i in range(5):          #adding 4 sheets
    wb1.sheets.add()
```

11.	Fetch a Single Column: How would you fetch the 'team1' column from the DataFrame?

```python
team=df['team1']
sheet1=wb1.sheets['Sheet1']
sheet1.range('A1').expand().value= team
sheet1.autofit()
```

22.	Fetch Multiple Columns: Can you fetch the 'team1', 'team2', and 'winner' columns together from the DataFrame?

```python
multiple_columns=df[['team1','team2','winner']]
sheet2=wb1.sheets['Sheet2']
sheet2.range('A1').expand().value= multiple_columns
sheet2.autofit()
```

13.	Fetch a Single Row: How would you fetch the 10th row of the DataFrame?

```python
row_specific=df.loc[10]
sheet3=wb1.sheets['Sheet3']
sheet3.range('A1').expand().value= row_specific
sheet3.autofit()
```

14.	Fetch Multiple Rows: Can you fetch rows 20 to 30 from the DataFrame?

```python
multiple_rows=df.loc[20:30]
sheet4=wb1.sheets['Sheet4']
sheet4.range('A1').expand().value= multiple_rows
sheet4.range('A:B').delete()
sheet4.autofit()
```

15.	Fetch Specific Rows and Columns: Fetch the 'city' and 'winner' columns for rows 50 to 60 from the DataFrame.

```python
specific_rows_columns=df.loc[50:60,['city','winner']]
sheet5=wb1.sheets['Sheet5']
sheet5.range('A1').expand().value= specific_rows_columns
sheet5.range('A:A').delete()
sheet5.autofit()
```
