```python
import xlwings as xw
import pandas as pd
```

```python
app=xw.App(visible=True)        #opens new workbook   #visible means changes are seen

wb1=app.books('Book1')
```

```python
df=pd.read_csv(r"D:\tasks\matches_csv.csv")    #load data 
```

```python
for i in range(7):          #adding 7 sheets
    wb1.sheets.add()
```

39.	Total Matches per Season: How many matches were played in each season?

```python
matches_per_season=df.groupby('season').size()
sheet1=wb1.sheets['Sheet1']
sheet1.range("A1").value = matches_per_season
sheet1.range('B1').value= 'total matches_per_season'
sheet1.autofit()
```

40.	Total Matches in Each City: How many matches were played in each city?

```python
matches_per_city=df.groupby('city').size()
sheet2=wb1.sheets['Sheet2']
sheet2.range("A1").value = matches_per_city
sheet2.range('B1').value= 'total matches_per_city'
sheet2.autofit()
```

41.	Total Wins per Team: What is the total number of wins for each team?

```python
total_teams=df.groupby('winner').size()
sheet3=wb1.sheets['Sheet3']
sheet3.range("A1").value = total_teams
sheet3.range('B1').value= 'total wins_per_team'
sheet3.autofit()
```

42.	Total Wins in Each Season: How many wins did each team secure in each season?

```python
wins_per_season_per_team=df.groupby(['winner','season']).size()
sheet4=wb1.sheets['Sheet4']
sheet4.range("A1").value = wins_per_season_per_team
sheet4.range('B1').value= 'total wins by each team in each season'
sheet4.autofit()
```

43.	Maximum Win by Runs per Season: What is the maximum margin of victory (in runs) in a match for each season?

```python
#sheet5.clear()
```

```python
max_win_by_runs = df.groupby('season')['win_by_runs'].max().reset_index()
sheet5=wb1.sheets['Sheet5']
sheet5.range("A1").options(index=False).value = max_win_by_runs
#sheet5.range('B1').value= 'max wins by runs of each team in each season'
sheet5.autofit()
```

44.	Filter Matches by Specific Teams: How would you filter the matches where either 'Mumbai Indians' or 'Chennai Super Kings' played as either 'team1' or 'team2'?

```python
filtered_matches = df[ (df['team1'].isin(['Mumbai Indians','Chennai Super Kings'])) | (df['team2'].isin(['Mumbai Indians','Chennai Super Kings'] ))  ]
sheet6=wb1.sheets['Sheet6']
sheet6.range("A1").options(expand='table',index=False).value = filtered_matches
sheet6.range('A:A').delete() 
sheet6.autofit()
```

45.	Filter Matches in Specific Cities: Can you filter the matches that took place in either 'Bangalore' or 'Delhi'?

```python
filtered_matches_cities = df[ (df['city'].isin(['Bangalore','Delhi'])) ]
sheet7=wb1.sheets['Sheet7']
sheet7.range("A1").options(expand='table',index=False).value = filtered_matches_cities
sheet7.range('A:A').delete() 
sheet7.autofit()
```

46.	Find Matches with Specific Umpires: How would you find the matches where either 'S Ravi' or 'HDPK Dharmasena' were either 'umpire1' or 'umpire2'?

```python
filter_match_umpire = df[ (df['umpire1'].isin(['S Ravi','HDPK Dharmasena' ])) | (df['umpire2'].isin(['S Ravi','HDPK Dharmasena' ] ))  ]
sheet8=wb1.sheets['Sheet8']
sheet8.range("A1").options(expand='table',index=False).value = filter_match_umpire
sheet8.range('A:A').delete() 
sheet8.autofit()
```
