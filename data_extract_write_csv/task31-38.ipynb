```python
import xlwings as xw
import pandas as pd
```

```python
app=xw.App(visible=True)        #opens new workbook   #visible means changes are seen

wb1=app.books('Book1')
```

```python
df=pd.read_csv(r"D:\tasks\matches_csv.csv")    #load data 
```

```python
for i in range(7):          #adding 7 sheets
    wb1.sheets.add()
```

31.	Sort by a Single Column: How would you sort the DataFrame based on the 'win_by_runs' column in ascending order?

```python
sort_asc=df.sort_values(by='win_by_runs',ascending=True)
sheet1=wb1.sheets['Sheet1']
#sheet1.range('A1').value='Mumbai_indians'
sheet1.range('A1').expand().value=sort_asc
sheet1.range('A:B').delete()
sheet1.autofit()
```

33.	Sort and Display Top N Rows: Sort the DataFrame by 'win_by_runs' in descending order and display the top 5 matches.

```python
sort_desc_rows=df.sort_values(by='win_by_runs',ascending=False).head() #head shows 5 entries by default
sheet3=wb1.sheets['Sheet3']
#sheet1.range('A1').value='Mumbai_indians'
#sheet3.range("A1").options(pd.DataFrame, expand='table', header=1,index=False).value=sort_desc_rows
#sheet3.range('A1').expand().header(1).value=sort_desc_rows
sheet3.range("A1").options(index=False).value = sort_desc_rows
sheet3.range('A:A').delete()
sheet3.autofit()
```

34.	Sort and Display Bottom N Rows: Sort the DataFrame by 'win_by_runs' in ascending order and display the bottom 10 matches.

```python
win_runs_sort=df.sort_values(by='win_by_runs').tail(10)
sheet4=wb1.sheets['Sheet4']
sheet4.range("A1").options(index=False).value = win_runs_sort
sheet4.range('A:A').delete()
sheet4.autofit()
```

35.	Drop Duplicates Based on a Single Column: How would you drop duplicate rows based on the 'team1' column?

```python
duplic=df.drop_duplicates(subset=['team1'])
sheet5=wb1.sheets['Sheet5']
sheet5.range("A1").options(index=False).value = duplic
sheet5.range('A:A').delete()
sheet5.autofit()
```

36.	Drop Duplicates Based on Multiple Columns: Can you drop duplicate rows based on both 'team1' and 'team2' columns?

```python
duplicates=df.drop_duplicates(subset=['team1','team2'])
sheet6=wb1.sheets['Sheet6']
sheet6.range("A1").options(index=False).value = duplicates
sheet6.range('A:A').delete()
sheet6.autofit()
```

37.	Drop Duplicates and Keep First Occurrence: Drop duplicate rows based on 'city' and keep the first occurrence for each unique city.

```python
city_dupl=df.drop_duplicates(subset=['city'],keep='first')    #keep=first is default even if we dont write
sheet7=wb1.sheets['Sheet7']
sheet7.range("A1").options(index=False).value = city_dupl
sheet7.range('A:A').delete()
sheet7.autofit()
```

38.	Drop Duplicates and Keep Last Occurrence: Drop duplicate rows based on 'city' and keep the last occurrence for each unique city.

```python
city_duplic=df.drop_duplicates(subset=['city'],keep='last')
sheet8=wb1.sheets['Sheet8']
sheet8.range("A1").options(index=False).value = city_duplic
sheet8.range('A:A').delete()
sheet8.autofit()
```
