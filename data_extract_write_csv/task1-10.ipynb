```python
import xlwings as xw
import pandas as pd
```

```python
df=pd.read_csv(r"D:\tasks\matches_csv.csv")

season = df[df["season"] == 2015]
```

```python
app=xw.App(visible=True)        #opens new workbook   #visible means changes are seen
```

```python
wb1=app.books('Book1')
```

```python
sheet1=wb1.sheets[0]
```

```python
for i in range(9):          #adding 10 sheets
    wb1.sheets.add()
```

1.	How can you filter the dataset to only show matches from a specific season?

```python
season = df[df["season"] == 2015]

sheet1.range('A1').value=season
sheet1.range("A:B").delete()   #deleting datframe index and file index
```

2.	Can you demonstrate how to filter the dataset to display matches held in a particular city?

```python
sheet2=wb1.sheets['Sheet2']
```

```python
city = df[df['city']=='Chennai']
sheet2.range('A1').value=city
sheet2.range('A:B').delete()
```

3.	How would you extract the information for matches where the toss winner chose to bat?

```python
toss=df[df['toss_decision']=='bat']
sheet3=wb1.sheets['Sheet3']
sheet3.range('A1').value= toss
sheet3.range('A:B').delete()
```

4.	Show the steps to filter the dataset to only display matches where the result was "tie."


```python
result=df[df['result']=='tie']
sheet4=wb1.sheets['Sheet4']
sheet4.range('A1').value= result
sheet4.range('A:B').delete()
```

5.	How can you extract matches where the margin of victory was by runs and it was more than 50 runs?

```python
victory_by_runs=df[df['win_by_runs']>=50]
sheet5=wb1.sheets['Sheet5']
sheet5.range('A1').value= victory_by_runs
sheet5.range('A:B').delete()
```

6.	Demonstrate how to filter the dataset to show matches where the DL method was applied.

```python
dl_method=df[df['dl_applied']==1]
sheet6=wb1.sheets['Sheet6']
sheet6.range('A1').value= dl_method
sheet6.range('A:B').delete()
```

7.	How would you filter the dataset to display matches where a specific team won?

```python
winner=df[(df['winner']=='Kolkata Knight Riders')]
sheet7=wb1.sheets['Sheet7']
sheet7.range('A1').value= winner
sheet7.range('A:B').delete()
```

8.	Show the steps to filter the dataset to only display matches where a certain team won by a margin of more than 30 runs.

```python
winner_with_margin=df[(df['winner']=='Rising Pune Supergiants') & (df['win_by_runs'] >= 30) ]
sheet8=wb1.sheets['Sheet8']
sheet8.range('A1').value= winner_with_margin
sheet8.range('A:B').delete()
```

9.	How can you extract matches where the 'player_of_match' was from a particular team?

```python
player_of_match=df[(df['winner']=='Sunrisers Hyderabad') & (df['player_of_match']=='Rashid Khan')]
sheet9=wb1.sheets['Sheet9']
sheet9.range('A1').value= player_of_match
sheet9.range('A:B').delete()
```

10.	Demonstrate how to filter the dataset to show matches where a specific umpire1 was officiating.

```python
specific_umpire1=df[df['umpire1']=='AK Chaudhary']
sheet10=wb1.sheets['Sheet10']
sheet10.range('A1').value= specific_umpire1
sheet10.range('A:B').delete()
```
