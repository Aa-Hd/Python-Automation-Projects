```python
import xlwings as xw
import pandas as pd
```

```python
app=xw.App(visible=True)        #opens new workbook   #visible means changes are seen

wb1=app.books('Book1')
```

```python
df=pd.read_csv(r"D:\tasks\matches_csv.csv")    #load data 
```

```python
for i in range(7):          #adding 7 sheets
    wb1.sheets.add()
```

16.	Minimum Value: What is the minimum value in the 'win_by_runs' column?

```python
min_in_column=df['win_by_runs'].min()
sheet1=wb1.sheets['Sheet1']
sheet1.range('A1').value='min_in_column'
sheet1.range('A2').expand().value=min_in_column
sheet1.autofit()
```

17.	Maximum Value: What is the maximum value in the 'win_by_wickets' column?

```python
max_in_column=df['win_by_wickets'].max()
sheet2=wb1.sheets['Sheet2']
sheet2.range('A1').value='max_in_column'
sheet2.range('A2').expand().value= max_in_column
sheet2.autofit()
```

18.	Count of Non-Null Values: How many non-null values are there in the 'city' column?

```python
non_null=df['city'].count()
sheet3=wb1.sheets['Sheet3']
sheet3.range('A1').value='non_null'
sheet3.range('A2').expand().value= non_null
sheet3.autofit()
```

19.	Count of Unique Values: How many unique teams are there in the 'team1' column?

```python
unique_in_col=df['team1'].nunique()
sheet4=wb1.sheets['Sheet4']
sheet4.range('A1').value='unique_in_col'
sheet4.range('A2').expand().value= unique_in_col
sheet4.autofit()
```

20.	Sum of a Column: Calculate the total runs scored by all teams in the 'win_by_runs' column.

```python
sum_column_values=df['win_by_runs'].sum()
sheet5=wb1.sheets['Sheet5']
sheet5.range('A1').value='sum_column_values'
sheet5.range('A2').expand().value= sum_column_values
sheet5.autofit()
```

21.	Mean of a Column: What is the mean value of the 'win_by_runs' column?

```python
mean_column=df['win_by_runs'].mean()
sheet6=wb1.sheets['Sheet6']
sheet6.range('A1').value='mean_column'
sheet6.range('A2').expand().value= mean_column
sheet6.autofit()
```

22.	Standard Deviation: Calculate the standard deviation of the 'win_by_runs' column.

```python
st_deviation_column=df['win_by_runs'].std().__round__(1)
sheet7=wb1.sheets['Sheet7']
sheet7.range('A1').value='st_deviation_column'
sheet7.range('A2').expand().value= st_deviation_column
sheet7.autofit()
```

23.	Count of Occurrences: How many times does the team 'Mumbai Indians' appear in the 'team1' column?

```python
count_occurences=(df['team1']=='Mumbai Indians').sum()
sheet8=wb1.sheets['Sheet8']
sheet8.range('A1').value='count_occurences'
sheet8.range('A2').expand().value= count_occurences
sheet8.autofit()
```

```python
wb1.close()
```
